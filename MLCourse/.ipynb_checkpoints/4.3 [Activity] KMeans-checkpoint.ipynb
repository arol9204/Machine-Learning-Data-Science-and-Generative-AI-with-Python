# K-Means Clustering Example

Let's make some fake data that includes people clustered by income and age, randomly:

```python
from numpy import random, array

#Create fake income/age clusters for N people in k clusters
def createClusteredData(N, k):
    random.seed(10)
    pointsPerCluster = float(N)/k
    X = []
    for i in range (k):
        incomeCentroid = random.uniform(20000.0, 200000.0)
        ageCentroid = random.uniform(20.0, 70.0)
        for j in range(int(pointsPerCluster)):
            X.append([random.normal(incomeCentroid, 10000.0), random.normal(ageCentroid, 2.0)])
    X = array(X)
    return X
```

We'll use k-means to rediscover these clusters in unsupervised learning:

```python
%matplotlib inline

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from numpy import random

data = createClusteredData(100, 5)

model = KMeans(n_clusters=5)

# Note I'm scaling the data to normalize it! Important for good results.
model = model.fit(scale(data))

# We can look at the clusters each data point was assigned to
print(model.labels_)

# And we'll visualize it:
plt.figure(figsize=(8, 6))
plt.scatter(data[:,0], data[:,1], c=model.labels_.astype(float))
plt.show()
```

### Part 2: What happens if you choose different values of K?

In this part what I'm going to do is run a for loop where we can run a k-means model with different numbers of centroids given by the for, and then visualize them to check which one fits the data the best.

```python
fig, axs = plt.subplots(2, 4, figsize=(20, 20), sharex='all', sharey='all')
for i in range(2, 10):
    model = KMeans(n_clusters = i)
    model = model.fit(scale(data))
    
    
    #ax.legend('k = {}'.format(i))
    #ax.legend(loc="upper left")
    
    axs[ (i//6) , ((i-2)%4) ].scatter(data[:,0], data[:, 1], c=model.labels_.astype(float))
    
plt.show()
```

# Concepts applied in this notebook:

- **Unsupervised Learning:** This is a type of machine learning that involves training models on data without labeled responses. Unlike supervised learning where models learn from input-output pairs, unsupervised learning focuses on finding patterns or intrinsic structures within the input data. Common tasks in unsupervised learning include clustering, dimensionality reduction, and association rule learning. Since there are no explicit correct answers or labels provided, the algorithm tries to organize the data in some way, such as grouping similar data points together or representing the data in a more compact and interpretable form. Unsupervised learning is particularly useful for exploratory analysis, discovering hidden patterns, or pre-processing data for further analysis.


- **K-Means:** K-means is a popular clustering algorithm used in unsupervised learning. The goal of k-means is to partition the data into K distinct clusters based on feature similarity. The basic algorithm involves choosing K initial centroids, then iteratively assigning each data point to the nearest centroid to form K clusters. After all data points have been assigned to clusters, the centroids are recalculated based on the current cluster memberships, and the process repeats until the centroids no longer change significantly. This results in a partitioning of the data space into Voronoi cells. K-means is widely used due to its simplicity and efficiency, although it requires the number of clusters to be specified in advance and can be sensitive to the initial choice of centroids.  


- **Scaled Data:** Scaling data is a preprocessing step in which the range of values of quantitative features in the dataset are adjusted or transformed. The purpose of scaling is to ensure that no variable dominates others purely because of their scale, allowing the model to converge more quickly and smoothly. Common scaling techniques include normalization (adjusting values to be between 0 and 1) or standardization (adjusting values so they have a mean of 0 and a standard deviation of 1). Scaling is particularly important in algorithms that calculate distances between data points, such as k-means or k-nearest neighbors, because it ensures that each feature contributes equally to the distance computations. Without scaling, features with larger ranges could disproportionately influence the results.
