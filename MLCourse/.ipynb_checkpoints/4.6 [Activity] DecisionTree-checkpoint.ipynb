# Decison Trees

First we'll load some fake data on past hires I made up. Note how we use pandas to convert a csv file into a DataFrame:

```python
import numpy as np
import pandas as pd
from sklearn import tree

input_file = "PastHires.csv"
df = pd.read_csv(input_file, header = 0)
```

```python
df.head()
```

scikit-learn needs everything to be numerical for decision trees to work. So, we'll map Y,N to 1,0 and levels of education to some scale of 0-2. In the real world, you'd need to think about how to deal with unexpected or missing data! By using map(), we know we'll get NaN for unexpected values.

```python
d = {'Y': 1, 'N': 0}
df['Hired'] = df['Hired'].map(d)
df['Employed?'] = df['Employed?'].map(d)
df['Top-tier school'] = df['Top-tier school'].map(d)
df['Interned'] = df['Interned'].map(d)

d = {'BS': 0, 'MS': 1, 'PhD': 2}
df['Level of Education'] = df['Level of Education'].map(d)
df.head()
```

Next we need to separate the features from the target column that we're trying to bulid a decision tree for.

```python
features = list(df.columns[:6])
features
```

Now actually construct the decision tree:

```python
y = df["Hired"]
X = df[features]
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X,y)
```

... and display it. Note you need to have pydotplus installed for this to work. (!pip install pydotplus)

To read this decision tree, each condition branches left for "true" and right for "false". When you end up at a value, the value array represents how many samples exist in each target value. So value = [0. 5.] mean there are 0 "no hires" and 5 "hires" by the tim we get to that point. value = [3. 0.] means 3 no-hires and 0 hires.

```python
from IPython.display import Image  
#from sklearn.externals.six import StringIO  
from six import StringIO
import pydotplus

dot_data = StringIO()  
tree.export_graphviz(clf, out_file=dot_data,  
                         feature_names=features)  
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())  
```

## Ensemble learning: using a random forest

We'll use a random forest of 10 decision trees to predict employment of specific candidate profiles:

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=10)
clf = clf.fit(X, y)

#Predict employment of an employed 10-year veteran
print (clf.predict([[10, 1, 4, 0, 0, 0]]))
#...and an unemployed 10-year veteran
print (clf.predict([[10, 0, 4, 0, 0, 0]]))
```

### Concepts applied in this notebook:

These concepts are key to understanding certain machine learning algorithms, especially within the context of classification and regression tasks. Let's explore each of them:


- **Decision Tree:** A decision tree is a flowchart-like tree structure where an internal node represents a feature(or attribute), the branch represents a decision rule, and each leaf node represents the outcome. The topmost node in a decision tree is known as the root node. It learns to partition on the basis of the attribute value. It partitions the tree in a recursive manner called recursive partitioning. This method is known as a simple, yet powerful classification and regression technique. Decision trees help in making decisions based on previous data (training data) in the form of a tree-structured graph, where each path from the root to a leaf represents a classification rule. The key advantage of decision trees is their ease of interpretation and understanding.


- **Random Forest:** A random forest is an ensemble learning technique which operates by constructing a multitude of decision trees at training time and outputting the class that is the mode of the classes (classification) or mean prediction (regression) of the individual trees. Random decision forests correct for decision trees' habit of overfitting to their training set. It works by creating multiple decision trees during training and aggregating their predictions for final output, which enhances the overall performance and accuracy of the model. Random forests can be used for both classification and regression tasks and are known for their high accuracy, robustness, and ease of use.


- **Entropy:** In the context of machine learning and information theory, entropy is a measure of the amount of uncertainty or randomness in a dataset. Specifically, in decision trees, entropy is used as a criterion to split the data. It helps to identify the feature that provides the most useful information for splitting the data in order to achieve the most homogeneous sub-nodes. Entropy is calculated for a dataset with two or more classes, and the goal of a decision tree is to reduce the entropy - or increase the purity - of the nodes with each split. The formula for entropy in a binary classification problem, for instance, can be represented as: \( - \sum p(x) \log_2 p(x) \), where \( p(x) \) is the probability of class \( x \). Lower entropy means less uncertainty, which in the context of decision trees, means that after the split, we have a purer subset.


These concepts are foundational in the field of machine learning, particularly in classification and regression tasks. Decision trees provide a simple yet effective method for making predictions and understanding decision paths, while random forests improve on the simplicity of decision trees by adding robustness and accuracy through ensemble learning. Entropy is a crucial measure in the process of building decision trees, guiding the selection of the features and splits that result in the most informative nodes.  


- **Ensemble learning**: is a powerful machine learning paradigm where multiple models (often called "weak learners") are trained to solve the same problem and combined in some way to improve the overall performance. This approach can lead to higher accuracy, better generalization to unseen data, and more robust models compared to any single model. Ensemble methods can be particularly effective because they exploit the diversity among the models to reduce errors due to bias and variance. Let's explore some specific types of ensemble learning:  

    - **Bagging**: Short for Bootstrap Aggregating, bagging is an ensemble learning technique designed to improve the stability and accuracy of machine learning algorithms. It works by creating multiple versions of a predictor model and training each one on a random subset of the training data. These subsets are created through bootstrap sampling, which means randomly selecting data points from the training set with replacement. The final prediction is made by aggregating the predictions of all the models, typically by taking the mean of the predictions for regression tasks or the majority vote for classification tasks. Random forests are a popular example of bagging applied to decision trees.  

    - **Boosting**: Boosting is an ensemble technique that aims to create a strong classifier from a number of weak classifiers. This is achieved by building a model from the training data, then creating a second model that attempts to correct the errors from the first model. Models are added sequentially until no further improvements can be made. Each model is given a weight based on its accuracy, and more weight is given to the models that predict incorrectly, making the algorithm focus on the harder cases in the training data. AdaBoost (Adaptive Boosting) and Gradient Boosting are well-known examples of boosting methods.  

    - **Bucket Models**: Bucket models, also known as "blending," involve training several different models and then combining their predictions. Each model in the "bucket" may be trained on the full dataset, but they can be entirely different types (e.g., decision trees, neural networks, SVMs). The idea is that by blending the strengths of various models, one can potentially achieve better performance than any single model could on its own. The final prediction could be a simple average of the predictions (for regression) or a majority vote/weighted vote (for classification).  

    - **Stacking**: Stacking, or stacked generalization, involves training a new model to combine the predictions of several other models. The original models are trained on the full dataset, and then a new model is trained to make a final prediction using the predictions of the other models as inputs. This second-level model can learn to correct the mistakes of the base models, leading to improved overall performance. Stacking is different from bagging and boosting in that it combines the models in a different space (the space of model predictions) and allows for the combination of very different types of models.  


Ensemble learning techniques like these are integral to many state-of-the-art machine learning systems, as they can significantly improve prediction accuracy on complex problems by combining the strengths of multiple models and mitigating their weaknesses.
